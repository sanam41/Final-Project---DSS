## Daraset overview
#### This dataset includes time series data generated by accelerometer and gyroscope sensors. It is collected with an iPhone 6s kept in the participant's front pocket using SensingKit which collects information from Core Motion framework on iOS devices. A total of 24 participants in a range of gender, age, weight, and height performed 6 activities in 15 trials in the same environment and conditions: downstairs, upstairs, walking, jogging, sitting, and standing. 

#### Here is an overview of one of the files: subject 1 - downstairs

```python
# Import libraries
import os
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/A_DeviceMotion_data/dws_1/sub_1.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
print(f'The sensors that have recorded signals are: \n {df.columns}')
```

```python
print(f'There are 24 subjects and each performed 15 trials. \n Each csv file has a shape of {df.shape}. \n Therefore, the size of the dataset is {df.shape[0] * 15} with {df.shape[1]} features.')
```

```python
df['gravity.x'].plot();
# Add labels and title
plt.xlabel('Data points')
plt.ylabel('gravity_x sensor')
plt.title('Time series plot')

# Show the plot
plt.show()
```
