## Human activity sensing using iPhone and Long short-term memory (LSTM) Recurrent Neural Network (RNN) architecture

This notebook shows the implementation of LSTM neural network architecture to Motion Sense dataset.

The goal was to compare it with the implementation of random forest.

Due to high demand of this approach for computational resources and dataset size, and the fact that random forest performed pretty well, this comparison demonstrates that it is always recommended to test much simpler algorithms to avoid the challenges of complicated approaches.

This notebook does not include hyperparameter tuning, such as the unit size of lstm layers, the choice of optimizer, loss function etc. It includes, 5-fold cross validation and model training and testing.



### Loading the data and merging in a single dataframe
The sensor data is collected by different participants, over different trials while they were performing different activities. 
Each trial is saved in a csv. Therefore, the first step is to merge the datasets and assure that the appropriate labels are added for accurate tracking.

```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import seaborn as sns
```

```python
# Looping through the folders to gather the information of all data files
file_path = 'data/A_DeviceMotion_data/'
folders = os.listdir(file_path)
files = {}
for f in folders:
    file = os.listdir(os.path.join(file_path, f))
    files[f] = file
                      
# Read the files containing the subject information
# Implementation of this information is beyond the scope of this project, but interesting insight can be derived from this dataset as well
subject_info = pd.read_csv('data/data_subjects_info.csv')
subject_info.rename(columns = {'code':'subject'}, inplace = True)

     
```

The folder names needs to be cleaned so that we can extract the activity type for each dataset and add it as an entry to a new column that determines whether the participants was jogging, sitting, running, etc.

```python
# renaming the columns:
folders_split = [f.split('_') for f in folders]
unique_activity = list(set([f[0] for f in folders_split])) # The unique type of activities are extracted here
# This list stores the activity type that will be added as a new column
new_col = []
for act in unique_activity:
    k = 1
    for i in range(len(folders_split)):
        if act == folders_split[i][0]:
            new_col.append(f'{act}_{k}')
            k += 1
```

```python
# create a dataframe and concat all data from different subjects and different activities: two columns are added that includes subject and the activity
dataset = []
labels = []
for key, value in files.items():
    subjects = os.listdir(os.path.join(file_path, key))
    for sub in subjects:
        data_sub = pd.read_csv(os.path.join(file_path, key, sub))
        data_sub = data_sub.drop(columns = 'Unnamed: 0')
        labels.append(key[:3])
        # data_sub['subject'] = f'{int(sub[4:-4])}'
        # data_sub['activity'] = f'{key[:3]}'
        dataset.append(data_sub)
        
```

## Model training

I have used these reference to build the lstm architecture:
https://machinelearningmastery.com/sequence-classification-lstm-recurrent-neural-networks-python-keras/
https://towardsdatascience.com/lstm-recurrent-neural-networks-how-to-teach-a-network-to-remember-the-past-55e54c2ff22e

To build the LSTM architecture I used keras functional API, and not the Sequential API. sequentual is more straightforward, but functional API provides more flexibility in contructing the architecture. This was my first time and I wanted to test and try this method.

```python
# Libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout
```

#### Model architecture
The stacking of the layers can become very complicated and optimizing the best architecture can take weeks.
Therefore, I decided to create a simple architectre and perform hyperparameter tuning on some of the parameters like the unit size of the LSTM layer (the size of the output) or the output size of the Dense layer and also dropout ratio.


```python
# This function creates our lstm model
def my_lstm(X_model,lstm1, dropout, lstm2, activation_f, optimizer_m, loss_m, metrics ):
    # The input layer defines the shape of the input data. 
    input_layer = Input(shape=(X_model.shape[1], X_model.shape[2])) 
    
    # The LSTM (Long Short-Term Memory) layer is a recurrent neural network layer used for sequence processing
    lstm_layer1 = LSTM(lstm1, return_sequences=True)(input_layer) # The last paranthesis shows the input to the layer
    
    # The dropout layer is used for regularization to prevent overfitting. It randomly zeroes out a fraction of input units during training
    dropout_layer = Dropout(dropout)(lstm_layer1)
    
    # A second lstm
    lstm_layer2 = LSTM(lstm2)(dropout_layer)
    
    # output layer with 6 outputs as the probability for each class
    output_layer = Dense(6, activation = activation_f)(lstm_layer2)

    # create the model
    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(optimizer= optimizer_m, loss=loss_m, metrics=[metrics])
    return model
```

### Cross-validation

```python
# Concatenate the arrays along the first axis
X_cv = np.concatenate((X_train_array, X_test_array), axis=0)
y_cv = np.concatenate((y_train_array, y_test_array), axis=0)
```

Cross validation was performed with 5 splits. The number of epochs was set to 400 so that calidation can be completed in a reasonable time

```python
from sklearn.model_selection import KFold
from tensorflow.keras.optimizers import Adam


# Initialize KFold with desired number of folds (e.g., 5-fold cross-validation)
n_splits = 5
kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Perform cross-validation
fold_accuracy = []  # List to store accuracy for each fold

for fold_idx, (train_idx, test_idx) in enumerate(kfold.split(X_cv, y_cv), 1):
    print(f"Fold {fold_idx}:")
    X_train_fold, X_test_fold = X_cv[train_idx], X_cv[test_idx]
    y_train_fold, y_test_fold = y_cv[train_idx], y_cv[test_idx]

    # Create and compile the LSTM model using the custom function
    model = my_lstm(X_train_fold, lstm1=128, dropout=0.2, lstm2=64, activation_f='softmax',
                    optimizer_m=Adam(learning_rate=0.001), loss_m='sparse_categorical_crossentropy',
                    metrics='accuracy')

    # Train the model on the training fold
    model.fit(X_train_fold, y_train_fold, epochs=400, batch_size=64, verbose=1)

    # Evaluate the model on the validation fold
    _, accuracy = model.evaluate(X_test_fold, y_test_fold)
    fold_accuracy.append(accuracy)

    print(f"Accuracy on validation fold: {accuracy}\n")

# Calculate and print the average accuracy across all folds
average_accuracy = np.mean(fold_accuracy)
print(f"Average accuracy across all folds: {average_accuracy}")
```

```python
# Train the lstm model on 80% and test on 20%
model = my_lstm(X_train_array, lstm1 = 128, dropout = 0.2, lstm2 = 64, 
                activation_f = 'softmax', 
               optimizer_m = Adam(learning_rate=0.001),
               loss_m = 'sparse_categorical_crossentropy',
                metrics = 'accuracy')
# Train the model
history = model.fit(X_train_array, y_train, epochs=1000, batch_size=64, validation_split = 0.2)
```

#### Model Evaluation

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt


# Evaluate the model on the test set
y_pred_probabilities = model.predict(X_test_array)
y_pred = np.argmax(y_pred_probabilities, axis=1)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot confusion matrix as a heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()


```
